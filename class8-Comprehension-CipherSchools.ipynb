```python
a = []
for i in range(5):
  a.append(i**2)

print(a)
```

```python
a = []
for i in range(5):
  temp = []
  for j in range(5):
    temp.append(j)
  a.append(temp)

print(a)
```

```python
a = []
for i in range(2):
  for j in range(2):
    a.append(j)
print(a)
```
